##### Calculating confirmed, recovered, death and mortality rate for global data 

```python
confirmed = con_global.iloc[:, 4 :]
death = de_global.iloc[:, 4:]
recover = rec_global.iloc[:, 4:]
```

```python
con_total = []
de_total = []
rec_total = []

#Lists to contain the per day value 
```

```python
#calculating the number of confirmed cases
for i in confirmed:
    con_total.append(confirmed[i].sum())
#calculating the totnal nummber of deaths 
for i in death:
    de_total.append(death[i].sum())

#calculating the number of recovered cases 
for i in recover:
    rec_total.append(recover[i].sum()) 

#calculating the aactive cases per day 


ac_total  =  [( con_total[i] - (de_total[i] + rec_total[i] ) ) for i in range(250)  ]
```

```python
con_total
```

```python
ac_total
```

```python
''''creating a dataframe of total case reports '''
#we need to create a dataframe from which we can directly plot using plotly 
world_total = pd.DataFrame(columns = ['Confirmed cases', 'Covid related Deaths',
                                      'Recovered cases', 'Active Covid cases', 'Dates'])
world_total['Confirmed cases'] = con_total
world_total['Covid related Deaths'] = de_total
world_total['Recovered cases'] = rec_total
world_total['Active Covid cases'] =  ac_total
world_total['Dates'] = dates
world_total['Dates'] = pd.to_datetime(world_total['Dates']) #converting date to datetime format 
world_total.set_index('Dates', inplace = True)
```

######  Daily Covid-19 graphs 

```python
#calculating the total number of confirmed cases 'perday':
confirmed_perday = []
deaths_perday = []
active_perday = []
recovered_perday = []
for i in range(249):
    confirmed_perday.append(world_total['Confirmed cases'].iloc[i+1] - world_total['Confirmed cases'].iloc[i] )
    deaths_perday.append(world_total['Covid related Deaths'].iloc[i+1] - world_total['Covid related Deaths'].iloc[i])
    recovered_perday.append(world_total['Recovered cases'].iloc[i+1] - world_total['Recovered cases'].iloc[i])
    active_perday.append(world_total['Active Covid cases'].iloc[i+1] - world_total['Active Covid cases'].iloc[i] )
    
```

```python
daily_update = pd.DataFrame(index = world_total.index[1:], columns = [ 'Confirmed', 'Deaths','Active','Recovered'])
daily_update['Confirmed'] = confirmed_perday
daily_update['Deaths'] = deaths_perday
daily_update['Recovered'] = recovered_perday
daily_update['Active'] = active_perday
```

```python
fig = px.line( data_frame = daily_update,  x = daily_update.index, y = 'Confirmed',
              title ='Covid 19 confirmed cases  (Daily)' )
fig.show()
```

```python
fig = px.line( data_frame = daily_update,  x = daily_update.index, y = 'Recovered',
              title ='Covid 19 Recovery cases (Daily)' )
fig.show()
```

```python
fig = px.line( data_frame = daily_update,  x = daily_update.index, y = 'Active',
              title ='Covid 19 Active cases (Daily)' )
fig.show()
'''
it is curious to notice that the daily active graph goes negative. After triple checking the data
and the code, I have come to the conclusion that a variety of reasons can lead to this 
peculiar graph :
a > Countries have reported mass recoveries/deaths, taking the number of active cases down on some
particular days. (We find a spike in recoveries/deaths, for example, recoveries on 6/8/20)
b> Countries have reported less confirmed cases on some particular days leading to a very minute increase 
in active cases or a steep drop (depends on countries).
c > Lack of data / misreporting of data could be another reason.
'''
```

```python
fig = px.line( data_frame = daily_update,  x = daily_update.index, y = 'Deaths',
              title ='Covid 19 Deaths (Daily)' )
fig.show()
```

```python
#Average values of Covid parameters
#We have data for 174 days 
confirm_average = []
death_average = []
active_average =  []
recovered_average = []
count = 0 
for  i in range(250):
    confirm_average.append( world_total['Confirmed cases'].iloc[i] / (i+1) )
    death_average.append( world_total['Covid related Deaths'].iloc[i] / (i+1))
    active_average.append( world_total['Active Covid cases'].iloc[i] / (i+1))
    recovered_average.append( world_total['Recovered cases'].iloc[i] / (i+1))

#Creating a Dataframe for Average Values     
Average = pd.DataFrame(index = world_total.index ,
                       columns = ['Confirmed cases (Avg)','Deaths(Avg)','Recovered cases(Avg)',
                                 'Active cases(Avg)'])
Average['Confirmed cases (Avg)'] = confirm_average
Average['Deaths(Avg)'] = death_average
Average['Recovered cases(Avg)'] = recovered_average
Average['Active cases(Avg)']  = active_average    
```

 ###### Plotting average COVID 19 confirmed cases, recoveries, active cases and deaths over time

```python
Country_Name = input("Enter the country name : ")

country_con = con_global[con_global['Country/Region'] == Country_Name ].T[4:]
country_rec = rec_global[rec_global['Country/Region'] == Country_Name].T[4:]
country_de = de_global[de_global['Country/Region'] == Country_Name].T[4:]
CountryDF  = pd.concat([country_con, country_rec ,country_de ], axis = 1)
    
CountryDF.columns = ['Confirmed cases','Recovered cases','Covid related Deaths' ]
CountryDF['Active Covid cases'] = [CountryDF['Confirmed cases'].iloc[i] - (CountryDF['Recovered cases'].iloc[i]+ CountryDF['Covid related Deaths'].iloc[i]) 
                                      for i in range(len(CountryDF.index))]
```
